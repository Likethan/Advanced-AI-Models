```python
# Install required libraries (optional)
!pip install -q tensorflow

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense

# Generate synthetic sine wave data
def generate_sine_data(seq_length=50, total=1000):
    x = np.arange(total)
    y = np.sin(0.02 * x) + 0.5 * np.random.randn(total)  # noisy sine wave
    X, Y = [], []
    for i in range(len(y) - seq_length):
        X.append(y[i:i + seq_length])
        Y.append(y[i + seq_length])
    return np.array(X), np.array(Y)

# Data preparation
seq_length = 50
X, Y = generate_sine_data(seq_length)
X = X.reshape(-1, seq_length, 1)
Y = Y.reshape(-1, 1)

# Split into train/test
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
Y_train, Y_test = Y[:split], Y[split:]

# Build RNN model
rnn_model = Sequential([
    SimpleRNN(50, activation='tanh', input_shape=(seq_length, 1)),
    Dense(1)
])
rnn_model.compile(optimizer='adam', loss='mse')
rnn_model.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=1)

# Build LSTM model
lstm_model = Sequential([
    LSTM(50, activation='tanh', input_shape=(seq_length, 1)),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=1)

# Evaluate and plot
pred_rnn = rnn_model.predict(X_test)
pred_lstm = lstm_model.predict(X_test)

plt.figure(figsize=(12, 5))
plt.plot(Y_test, label='True')
plt.plot(pred_rnn, label='RNN Prediction')
plt.plot(pred_lstm, label='LSTM Prediction')
plt.legend()
plt.title("Time-Series Prediction")
plt.show()
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence

# Load IMDB dataset
max_features = 10000  # top 10k words
maxlen = 200  # max sequence length

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)

# RNN model
rnn_model = Sequential([
    tf.keras.layers.Embedding(max_features, 128),
    SimpleRNN(64),
    Dense(1, activation='sigmoid')
])
rnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
rnn_model.fit(x_train, y_train, epochs=3, batch_size=64, validation_split=0.2)

# LSTM model
lstm_model = Sequential([
    tf.keras.layers.Embedding(max_features, 128),
    LSTM(64),
    Dense(1, activation='sigmoid')
])
lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
lstm_model.fit(x_train, y_train, epochs=3, batch_size=64, validation_split=0.2)

# Evaluate
rnn_acc = rnn_model.evaluate(x_test, y_test)[1]
lstm_acc = lstm_model.evaluate(x_test, y_test)[1]
print(f"RNN Test Accuracy: {rnn_acc:.4f}")
print(f"LSTM Test Accuracy: {lstm_acc:.4f}")
```
